```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("processed_data/processed_data.csv")
df['datetime'] = pd.to_datetime(df['datetime'])

# Filtrer les données pour inclure uniquement celles du mois de mai
df_may = df[df['datetime'].dt.month == 5]
df_may
```

```python
# Calculer le score cumulé de sentiment pour chaque date
df_may['cumulative_sentiment'] = df_may['financial_sentiment'].cumsum()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_may['datetime'], df_may['cumulative_sentiment'], marker='o')
plt.title('Évolution du sentiment financier pour le gaz en France au mois de mai')
plt.xlabel('Date')
plt.ylabel('Score de sentiment cumulé')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
